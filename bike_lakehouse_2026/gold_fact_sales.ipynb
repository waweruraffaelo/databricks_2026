```python
query = """
SELECT
cs.order_number,
cp.product_key,
cu.customer_id,
cs.order_date,
cs.ship_date,
cs.sales_amount,
cs.quantity,
cs.price
FROM workspace.silver.crm_sales cs
LEFT JOIN workspace.silver.crm_products cp
ON cs.product_number = cp.product_key
LEFT JOIN workspace.silver.crm_customers cu
ON cs.customer_id = cu.customer_id

"""

df = spark.sql(query)
```

```python
df.limit(3).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
