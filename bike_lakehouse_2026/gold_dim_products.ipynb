```python
query = """

SELECT
ROW_NUMBER() OVER(ORDER BY cp.start_date, cp.product_key) AS product_number,
cp.product_id,
cp.product_key,
cp.product_name,
cp.category_id,
pc.category,
pc.subcategory,
pc.maintenance_flag,
cp.product_line,
cp.start_date
FROM workspace.silver.crm_products cp
LEFT JOIN workspace.silver.erp_product_category pc
ON cp.category_id = pc.category_id
WHERE cp.end_date IS NULL
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

# Write  The product table to Gold Layer 

```python
df.write.mode("overwrite").format('delta').saveAsTable("workspace.gold.dim_products")
```
